```python
import os
import glob
import pandas as pd
```

```python
def get_files(filepath):
    all_files = []
    for root, dirs, files in os.walk(filepath):
        files = glob.glob(os.path.join(root,'*.json'))
        for f in files :
            all_files.append(os.path.abspath(f))
    
    return all_files
```

# All logs in one table

```python
df_logs = pd.concat([pd.read_json(cur_file, lines=True) for cur_file in get_files('./../data/log_data/')], axis = 0)
df_logs = df_logs[df_logs['page'] == 'NextSong']
df_logs.head()
```

# All songs in one table

```python
df_songs = pd.concat([pd.read_json(cur_file, lines=True) for cur_file in get_files('./../data/song_data/')], axis = 0)
df_songs.head()
```

# Very few records to mach from logs with songs

```python
pd.merge(
    left=df_logs[['song','artist','length']], 
    right= df_songs[['title','artist_name','song_id','artist_id','duration']], 
    left_on=['song','artist'], 
    right_on=['title','artist_name']
)
```
